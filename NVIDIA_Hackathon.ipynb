```python
keras = tf.keras

# Scaling the features
Xarray = X.values  # Get the values from X
X_scaled = (Xarray - Xarray.mean(axis=0)) / Xarray.std(axis=0)  # Standard scaling

# Scaling the target variable
y_array = train_data['y'].values  # Get the values from y
y_scaled = (y_array - y_array.mean()) / y_array.std()  # Standard scaling for y

# Split the scaled data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```
